```python
import pandas as pd
#NaN missing values
var=pd.read_csv("https://raw.githubusercontent.com/YBI-Foundation/Dataset/refs/heads/main/Diabetes%20Missing%20Data.csv")
var
```

```python
#now i want to DROP missing data
var.dropna() #it is dropped along ROW
```

```python
#we can also drop along column
var.dropna(axis=1) #1 for column
```

```python
var.dropna(axis=0) #0 for row
```

```python
#if my whole ROW is empty.. how to remove
var.dropna(how='any')
```

```python
 var.dropna(how='all') #here NaN's are there but only the mepty row is removed
```

```python
#remove null value along column
var.dropna(subset=['Serum_Insulin']) 
#here  null values from Serum_Insulin is removed
```

```python
import pandas as pd
var=pd.read_csv("https://raw.githubusercontent.com/YBI-Foundation/Dataset/refs/heads/main/Diabetes%20Missing%20Data.csv")
var.dropna(inplace=True)
var # null values are deleted and a new dataset is created
```

```python
import pandas as pd
var=pd.read_csv("https://raw.githubusercontent.com/YBI-Foundation/Dataset/refs/heads/main/Diabetes%20Missing%20Data.csv")
var.dropna(thresh=1) #jisme single null value woh DROP
var
```

```python
import pandas as pd
var=pd.read_csv("https://raw.githubusercontent.com/YBI-Foundation/Dataset/refs/heads/main/Diabetes%20Missing%20Data.csv")
var.dropna(thresh=2) #jisme 2 null value woh DROP
var
```

# fillna

```python
#it is used to fill values  
import pandas as pd
var=pd.read_csv("https://raw.githubusercontent.com/YBI-Foundation/Dataset/refs/heads/main/Diabetes%20Missing%20Data.csv")
var.fillna('python') #fill the empty spaces by python
```

```python
#to fill data in a particular column
var.fillna({'Serum_Insulin':'python','Skin_Fold':'c'})
```

```python
#to fill null value with the previous data
var.fillna(method="ffill")  #forward fill
```

```python
#backward filling
var.fillna(method="bfill")
```

```python
#fill value along axis
var.fillna(method="bfill",axis=0)#fill along row
 
```

```python
var.fillna(method="bfill",axis=1)#fill along column
 
```

```python
var.fillna('python',limit=2) #yahan first 2 Nan's of a column get filled by 'python'
```

# Replace

```python
#replace using regex
v=pd.read_csv('https://raw.githubusercontent.com/codeforamerica/ohana-api/refs/heads/master/data/sample-csv/addresses.csv')
v.replace('[A-Za-z]','python',regex=True)
```

```python
v=pd.read_csv('https://raw.githubusercontent.com/codeforamerica/ohana-api/refs/heads/master/data/sample-csv/addresses.csv')
v.replace('[A-Z]','python',regex=True) #all capitals gets converted to python
```

# interpolate

```python
#automatically fills data with the aboove observationss
var=pd.read_csv("https://raw.githubusercontent.com/YBI-Foundation/Dataset/refs/heads/main/Diabetes%20Missing%20Data.csv")
var
```

```python
var.interpolate() 
#interpolate only works with no's
#not alphabets (they remain NaN)
```

```python
var.interpolate(method='linear')  #data is filled linearly
```

```python
var.interpolate(method='linear',axis=0)# 0 for row
```

```python
var.interpolate(method='linear',axis=1)# 1 for column
#for column it may show error
#if data types are DIFFERENT (bagal bagal)
```

# Merge

```python
import pandas as pd
var1=pd.DataFrame({'A':[1,2,3,4],'B':[5,6,7,8]})
var2=pd.DataFrame({'A':[1,2,3,4],'c':[11,12,123,14]})
pd.merge(var1,var2,on='A')
#both the data frames should have some common part here A
#while merging (same id's)
```

```python
#if we want c before b in table
import pandas as pd
var1=pd.DataFrame({'A':[1,2,3,4],'B':[5,6,7,8]})
var2=pd.DataFrame({'A':[1,2,3,4],'C':[11,12,123,14]})
pd.merge(var2,var1,on='A')
```

```python
#if we change one of the values from the common a field,
# that is not reflected in table (WITH COMMON DATAFRAMES)
import pandas as pd
var1=pd.DataFrame({'A':[1,2,3,4],'B':[5,6,7,8]})
var2=pd.DataFrame({'A':[1,2,3,9],'C':[11,12,123,14]})
pd.merge(var2,var1,on='A')
```

```python
#how-> we can get the missing data in the dataframe
import pandas as pd
var1=pd.DataFrame({'A':[1,2,3,4],'B':[5,6,7,8]})
var2=pd.DataFrame({'A':[1,2,3,9],'C':[11,12,123,14]})
pd.merge(var1,var1,how="inner") #inner me no change
```

```python
import pandas as pd
var1=pd.DataFrame({'A':[1,2,3,4],'B':[5,6,7,8]})
var2=pd.DataFrame({'A':[1,2,3,9],'C':[11,12,123,14]})
pd.merge(var1,var2,how="left")    #inner me no change
```

```python
import pandas as pd
var1=pd.DataFrame({'A':[1,2,3,4],'B':[5,6,7,8]})
var2=pd.DataFrame({'A':[1,2,3,9],'C':[11,12,123,14]})
pd.merge(var1,var2,how="right")
```

```python
import pandas as pd
var1=pd.DataFrame({'A':[1,2,3,4],'B':[5,6,7,8]})
var2=pd.DataFrame({'A':[1,2,3,9],'C':[11,12,123,14]})
pd.merge(var1,var2,how="outer")
```

```python
import pandas as pd
var1=pd.DataFrame({'A':[1,2,3,4],'B':[5,6,7,8]})
var2=pd.DataFrame({'A':[1,2,3,9],'C':[11,12,123,14]})
pd.merge(var1,var2,how="outer",indicator=True)
```

```python
import pandas as pd
var1=pd.DataFrame({'A':[1,2,3,4],'B':[5,6,7,8]})
var2=pd.DataFrame({'A':[1,2,3,9],'C':[11,12,123,14]})
pd.merge(var1,var2,left_index=True,right_index=True)
```

```python
import pandas as pd
var1=pd.DataFrame({'A':[1,2,3,4],'B':[5,6,7,8]})
var2=pd.DataFrame({'A':[1,2,3,5],'C':[11,12,123,14]})
pd.merge(var1,var2,left_index=True,right_index=True,suffixes=("name","python"))
```

# concat

```python
#concat series
sr1=pd.Series([1,2,3,4])
sr2=pd.Series([11,21,31,41])
pd.concat([sr1,sr2])
```

```python
#concat dataframe
d1=pd.DataFrame([1,2,3,4])
d2=pd.DataFrame([11,21,31,41])
pd.concat([d1,d2])
#it does not takes common and merge
```

```python
#concat dataframe -> and put it across column
#merging across column
d1=pd.DataFrame([1,2,3,4])
d2=pd.DataFrame([11,21,31,41])
pd.concat([d1,d2],axis=1)
```

```python
import pandas as pd
var1=pd.DataFrame({'A':[1,2,3,4],'B':[11,12,13,14]})
var2=pd.DataFrame({'A':[1,2],'c':[21,22]})
pd.concat([var1,var2],axis=1,join="outer")
```

```python
import pandas as pd
var1=pd.DataFrame({'A':[1,2,3,4],'B':[11,12,13,14]})
var2=pd.DataFrame({'A':[1,2],'c':[21,22]})
pd.concat([var1,var2],axis=1,join="inner")#shows all missing data
```

```python
import pandas as pd
var1=pd.DataFrame({'A':[1,2,3,4],'B':[5,6,7,8]})
var2=pd.DataFrame({'A':[1,2,3,4],'B1':[11,12,13,14]})
pd.concat([var1,var2],keys=['var1','var2'],axis=0)
```
