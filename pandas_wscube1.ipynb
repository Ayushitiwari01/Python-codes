DATA STRUCTURE IN PYTHON PANDAS

1> SERIES DATA STRUCTURE

```python
#series data structure
import pandas as pd
```

```python
x=[1,2,3,4,5,6]
var=pd.Series(x)
print(var)
print(type(var))
print(var[2])#at index 2 -> val is 3
```

```python
#changing index values
x=[1,2,3,6]
var=pd.Series(x,index=['a','b','c','d'])
print(var)
```

```python
#changing data type
x=[1,2,3,6]
var=pd.Series(x,index=['a','b','c','d'],dtype='float')
print(var)
```

```python
#giving name:
x=[1,2,3,6]
var=pd.Series(x,index=['a','b','c','d'],dtype='float',name='pipi')
print(var)
```

```python
#now we will be using dictionary
dic={'name':['python','c','java','c++'],'por':[12,13,14,15]}
var=pd.Series(dic)
print(var)
#we get data type obect bcz we have mized data
#some is int ..some is string
```

```python
#making series using single data
s=pd.Series(12)
print(s)
print(type(s))
```

```python
s=pd.Series(12,index=[1,2,3,4,5,6,7])
print(s)
print(type(s))
```

```python
s1=pd.Series(12,index=[1,2,3,4,5,6,7])
s2=pd.Series(12,index=[1,2,3,4])
print(s1+s2)
#why do we use series when we have numpy arrays
#in series any operation we want .. is performed
#in series-> till 4 we have same data-> so can ADD ->after that NULL
#in numpy->there is broadcast error (as both have diff sizes)
# NaN-> not a number
```

ARITHMETIC OPERATION IN PANDAS

```python
import pandas as pd
var = pd.DataFrame({'a':[1,2,3,4],'b':[4,5,6,7]})
var
```

```python
#add
var['c']=var['a']+var['b']
var
```

```python
#sub
var['c']=var['a']-var['b']
var
```

```python
#mul
var['c']=var['a']*var['b']
var
```

```python
#div
var['c']=var['a']/var['b']
var
```

```python
var = pd.DataFrame({'a':[10,20,30,4],'b':[4,52,56,7]})
var
```

```python
#logical
var['python1']=var['a']<=20
var['python2']=var['b']<=20
var
```
